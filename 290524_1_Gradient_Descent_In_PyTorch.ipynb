```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = torch.tensor([0., 0.], requires_grad=True)
```

```python
alpha = 0.1
x = torch.tensor([0.0, 0.0], requires_grad=True)

function_values = []
for step in range(100):
    function = (1/ (1 + torch.exp(-x))).sum()
    function_values.append(function.item())
    print(f'step {step}: {function.item()}')
    function.backward()
    x.data = x - alpha * x.grad
    x.grad.zero_()
```

```python
plt.plot(function_values)
```
