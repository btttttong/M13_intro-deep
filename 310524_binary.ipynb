```python
import numpy as np

true = 0.9
pred = np.linspace(0.001, 0.999, 1000)
SE = (true - pred) ** 2
loss = SE

BCE = -true * np.log(pred) - (1 - true) * np.log(1 - pred)

import matplotlib.pyplot as plt
plt.scatter(pred, BCE, s=0.1)
plt.scatter(pred, SE, s=0.1)
print(BCE.min(), pred[np.argmin(BCE)])
```
